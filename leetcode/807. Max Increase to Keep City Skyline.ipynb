```go
package main

import (
	"fmt"
)

func get_min_value(v []int) int {
	e := v[0]
	for _, v := range v {
		if v < e {
			e = v
		}
	}
	return e
}

func get_max_value(v []int) int {
	e := v[0]
	for _, v := range v {
		if v > e {
			e = v
		}
	}
	return e
}

func maxIncreaseKeepingSkyline(grid [][]int) int {
	//3:47
	row_skyline := make([]int, 0)
	column_skyline := make([]int, 0)
	for _, row := range grid {
		row_skyline = append(row_skyline, get_max_value(row[:]))
	}
	for i := 0; i < len(grid); i++ {
		column := make([]int, 0)
		for j := 0; j < len(grid[0]); j++ {
			column = append(column, grid[j][i])
		}
		column_skyline = append(column_skyline, get_max_value(column[:]))
	}
	result := 0
	for y, row := range grid {
		for x, value := range row {
			arr := [2]int{row_skyline[y], column_skyline[x]}
			target := get_min_value(arr[:])
			result += target - value
		}
	}
	return result
	//4:03
	//debug from 4:30 to 4:32
}
```
