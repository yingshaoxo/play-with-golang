```go
package main

import (
	"fmt"
	"strconv"
)

func get_permutations(arr []int) [][]int {
	//copied
	var helper func([]int, int)
	res := [][]int{}

	helper = func(arr []int, n int) {
		if n == 1 {
			tmp := make([]int, len(arr))
			copy(tmp, arr)
			res = append(res, tmp)
		} else {
			for i := 0; i < n; i++ {
				helper(arr, n-1)
				if n%2 == 1 {
					tmp := arr[i]
					arr[i] = arr[n-1]
					arr[n-1] = tmp
				} else {
					tmp := arr[0]
					arr[0] = arr[n-1]
					arr[n-1] = tmp
				}
			}
		}
	}
	helper(arr, len(arr))
	return res
}

func format_the_time(hour int, minutes int) string {
	h := strconv.Itoa(hour)
	m := strconv.Itoa(minutes)
	if len(h) == 1 {
		h = "0" + h
	}
	if len(m) == 1 {
		m = "0" + m
	}
	return fmt.Sprintf("%s:%s", h, m)
}

func largestTimeFromDigits(arr []int) string {
	//9:29
	permutations := get_permutations(arr)
	min_length := -1
	the_time := make([]int, 2)
	for i := 0; i < len(permutations); i++ {
		a := permutations[i][0]
		b := permutations[i][1]
		c := permutations[i][2]
		d := permutations[i][3]
		hour := a*10 + b
		minutes := c*10 + d
		if ((0 <= hour) && (hour <= 23)) && ((0 <= minutes) && (minutes <= 59)) {
			length := hour*60 + minutes
			if length > min_length {
				min_length = length
				the_time[0] = hour
				the_time[1] = minutes
			}
		}
	}
	if min_length == -1 {
		return ""
	}
	return format_the_time(the_time[0], the_time[1])
	//9:48
}
```
