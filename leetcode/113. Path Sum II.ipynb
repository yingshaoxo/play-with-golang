```go
/**
 * Definition for a binary tree node.
 * type TreeNode struct {
 *     Val int
 *     Left *TreeNode
 *     Right *TreeNode
 * }
 */
 
import (
    "fmt"
)


var target int
var paths [][]int


func mySum(array []int) int {
	result := 0
	for _, v := range array {
		result += v
	}
	return result
}

func travel(node *TreeNode, l []int) {
	if node == nil {
		return
	} else {
		l = append(l, node.Val)
        if (node.Left == nil) && (node.Right == nil) {
			if mySum(l) == target {
                paths = append(paths, l[:])
                fmt.Printf("%v \n", l)
            }
        }
        length := len(l)
        travel(node.Left, l[:])
        travel(node.Right, l[:length])
	}
}

func pathSum(root *TreeNode, sum int) [][]int {
    paths = make([][]int, 0)
	target = sum
    travel(root, make([]int, 0))
    return paths
}
```
